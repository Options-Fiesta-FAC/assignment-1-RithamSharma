```python
import numpy as np
import matplotlib.pyplot as plt

strike_price = float(input("Enter the Strike Price: "))
premium = float(input("Enter the Option Premium: "))
spot_start = float(input("Enter the Starting Spot Price: "))
spot_end = float(input("Enter the Ending Spot Price: "))

def payoff_long_call(strike, premium, spot_range):
    return np.maximum(spot_range - strike, 0) - premium

def payoff_short_call(strike, premium, spot_range):
    return premium - np.maximum(spot_range - strike, 0)

def payoff_long_put(strike, premium, spot_range):
    return np.maximum(strike - spot_range, 0) - premium

def payoff_short_put(strike, premium, spot_range):
    return premium - np.maximum(strike - spot_range, 0)

spot_prices = np.linspace(spot_start, spot_end, 200)

long_call = payoff_long_call(strike_price, premium, spot_prices)
short_call = payoff_short_call(strike_price, premium, spot_prices)
long_put = payoff_long_put(strike_price, premium, spot_prices)
short_put = payoff_short_put(strike_price, premium, spot_prices)

plt.figure(figsize=(14, 8))
plt.plot(spot_prices, long_call, 'b-', label='Long Call')
plt.plot(spot_prices, short_call, 'r--', label='Short Call')
plt.plot(spot_prices, long_put, 'g-.', label='Long Put')
plt.plot(spot_prices, short_put, 'm:', label='Short Put')

plt.title("Option Payoff Diagrams", fontsize=16)
plt.xlabel("Spot Price", fontsize=14)
plt.ylabel("Profit / Loss", fontsize=14)
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```
